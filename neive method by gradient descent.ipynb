```python
m=1600
n=2048
X=data

y=Y


# Load the dataset (X: features, y: target variable)


# Normalize the feature values
X_normalized = (X - np.mean(X, axis=0)) / np.std(X, axis=0)

# Add a column of ones for the bias term
X_normalized = np.hstack((np.ones((X_normalized.shape[0], 1)), X_normalized))

# Initialize model parameters
num_features = X_normalized.shape[1]
theta = np.zeros(num_features)
alpha = 0.01
num_iterations = 1000

# Define the cost function (mean squared error)
def cost_function(X, y, theta):
    m = len(y)
    predictions = np.dot(X, theta)
    error = predictions - y
    cost = np.sum(error ** 2) / (2 * m)
    return cost

# Perform gradient descent
def gradient_descent(X, y, theta, alpha, num_iterations):
    m = len(y)
    cost_history = []
    for _ in range(num_iterations):
        predictions = np.dot(X, theta)
        error = predictions - y
        gradient = np.dot(X.T, error) / m
        theta -= alpha * gradient
        cost = cost_function(X, y, theta)
        cost_history.append(cost)
    return theta, cost_history
optimized_theta, cost_history = gradient_descent(X_normalized, y, theta, alpha, num_iterations)
```

```python
def HT( v, k ):
    t = np.zeros_like( v )
    if k < 1:
        return t
    else:
        ind = np.argsort( abs( v ) )[ -k: ]
        t[ ind ] = v[ ind ]
        return t
```
